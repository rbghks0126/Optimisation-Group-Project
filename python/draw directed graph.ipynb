```python
text = """
Provider 2 provides service
Provider 2 comes back to 14HS at 8.0hr.
1-7
7-1
travel length:3

Provider 3 provides service
Provider 3 comes back to 14HS at 7.0hr.
1-6
6-1
travel length:2

Provider 4 provides service
Provider 4 comes back to 14HS at 8.0hr.
1-3
2-1
3-2
travel length:5

Provider 5 provides service
Provider 5 comes back to 14HS at 8.0hr.
1-4
4-5
5-1
travel length:3"""
```

```python
import re
ptn = re.compile("[\d]+-[\d]")
routes = ptn.findall(text)
edges = []
for each in routes:
    edges+=[(each[0],each[2])]

import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph(directed=True)
G.add_edges_from(edges)
options = {
    'node_size': 400,
    'width': 1,
    'arrowstyle': '-|>',
    'arrowsize': 12,
}
nx.draw_networkx(G, arrows=True, **options)
plt.show()
```
