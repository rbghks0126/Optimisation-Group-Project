```python
text = """

1-2
2-7
7-1


1-6
6-1


1-4
4-8
8-1


1-10
10-3
3-1

1-9
9-1

1-11
11-5
5-1
"""
```

```python
import re
ptn = re.compile("[\d]+-[\d]+")
routes = ptn.findall(text)
edges = []
for each in routes:
    idx = each.index('-')
    edges+=[(each[:idx],each[idx+1:])]

import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph(directed=True)
G.add_edges_from(edges)
options = {
    'node_size': 400,
    'width': 1,
    'arrowstyle': '-|>',
    'arrowsize': 12,
    'font_size': 8,
}
nx.draw_networkx(G, arrows=True, **options)
plt.savefig("part1_10clients", bbox_inches='tight', dpi=300)
plt.show()
```
